```python
import matplotlib.pyplot as pp
import numpy as np
```

```python
# Trig function
def function1(x):

    return 100*np.sin(x), 100*np.cos(x)

# Polynomial
def function2(x):

    return x**3+x**2, 3*x**2-2*x

# Linear
def function3(x):

    return 10*x, np.ones(x.size)*10.0
```

```python
def forwardFD(x,y):

    dy = []
    xn = []
    for i in range(x.size-1):
        dy.append((y[i+1]-y[i])/(x[i+1]-x[i]))
        xn.append((x[i+1]+x[i])*0.5)

    return np.array(xn), np.array(dy)

def centeredFD(x,y):

    dy = []
    xn = []
    for i in range(1, x.size-1):
        dy.append((y[i+1]-y[i-1])/(x[i+1]-x[i-1]))
        xn.append(x[i])

    return np.array(xn), np.array(dy)

def reversedFD(x,y):

    dy = []
    xn = []
    for i in range(1,x.size):
        dy.append((y[i]-y[i-1])/(x[i]-x[i-1]))
        xn.append((x[i]+x[i-1])*0.5)
    
    return np.array(xn), np.array(dy)
```

```python
# Make an array of values and compute the derivatives for each function.
step_size = 0.5
x = np.arange(-5,5+step_size,step_size)
xc = x#np.arange(-5,5+step_size*0.5,step_size*0.5)

# Make the plots
fig, (ax1, ax2, ax3) = pp.subplots(nrows=3, figsize=(10,8))

# Linear function
y, dy = function3(x)
yc, dyc = function3(xc)
x2, dy2 = forwardFD(x,y)
x3, dy3 = centeredFD(xc,yc)
x4, dy4 = reversedFD(x,y)
ax1.plot(x, y, color='black', linestyle='--', label='Function')
ax1.plot(x, dy, color='black', label='Derivative')
ax1.plot(x2, dy2, color='firebrick', label='Forward FD', marker='s')
ax1.plot(x3, dy3, color='dodgerblue', label='Centered FD', marker='^')
ax1.plot(x4, dy4, color='forestgreen', label='Reversed FD')

# Polynomial function
y, dy = function2(x)
yc, dyc = function2(xc)
x2, dy2 = forwardFD(x,y)
x3, dy3 = centeredFD(xc,yc)
x4, dy4 = reversedFD(x,y)
ax2.plot(x, y, color='black', linestyle='--')
ax2.plot(x, dy, color='black', label='Derivative')
ax2.plot(x2, dy2, color='firebrick', label='Forward FD', marker='s')
ax2.plot(x3, dy3, color='dodgerblue', label='Centered FD', marker='^')
ax2.plot(x4, dy4, color='forestgreen', label='Reversed FD')

# Trig function
y, dy = function1(x)
yc, dyc = function1(xc)
x2, dy2 = forwardFD(x,y)
x3, dy3 = centeredFD(xc,yc)
x4, dy4 = reversedFD(x,y)
ax3.plot(x, y, color='black', linestyle='--')
ax3.plot(x, dy, color='black', label='Derivative')
ax3.plot(x2, dy2, color='firebrick', label='Forward FD', marker='s')
ax3.plot(x3, dy3, color='dodgerblue', label='Centered FD', marker='^')
ax3.plot(x4, dy4, color='forestgreen', label='Reversed FD')

# Decorations
ax1.legend(loc='upper left')

pp.show()
```
